<h3> Univarijatni odabir </h3>

Makroekonomski indikatori su znacajke koje pocinju sa 'ind'. Imena tih značajki su izmijenjena u bilježnici "data_preparation". 
Prava imena su:
1. ind1-poc = GDP per capita (in EUR)  -------> ovo se ne koristi, izabran je ind2-poc
2. ind2-poc = GDP - year-on-year rate of growth (in %, constant prices) 
3. ind3-poc = Average CPI year-on-year inflation rate 
4. ind4-poc = Current account balance (as of % GDP) 
5. ind5-poc = Exports of goods and services (as of % GDP)
6. ind6-poc = Imports of goods and services (as of % GDP) 
7. ind7-poc = External debt (as of % GDP) 
8. ind8-poc = Average exchange rate (HRK : 1 EUR)
9. ind9-poc = Average exchange rate (HRK : 1 USD)
10. ind10-poc = Consolidated general government net lending (+)/borrowing (-) (as % of GDP)
11. ind11-poc = General government debt (as % of GDP)
12. ind12-poc = Employment rate (ILO, persons above 15 years of age)

```python
from sklearn.ensemble import RandomForestClassifier
train, test = train_test_split(df, test_size=0.20, random_state=42)

forest = RandomForestClassifier(n_estimators=1000,random_state=0,n_jobs=-1) # Promjeni n_jobs za paralelizaciju!

X_train = train.drop('PRIJEVREMENI_RASKID',axis=1)
y_train = train['PRIJEVREMENI_RASKID']

forest.fit(X_train, y_train)
importances = forest.feature_importances_
indices = np.argsort(importances)[::-1] # Vrati indekse koji sortiraju polje (obratno)

plt.figure(figsize=(7,5))

plt.plot(range(X_train.shape[1]), importances[indices], 'o', color='gray')

plt.title('Odabir znacajki uz pomoc slucajnih suma')
plt.ylabel('vaznost znacajke')
plt.xlim([-1, X_train.shape[1]])
plt.xticks(range(X_train.shape[1]), train.drop('PRIJEVREMENI_RASKID',axis=1).columns.values[indices], rotation=90)
plt.tight_layout()
plt.grid(True,linestyle='--')
plt.show()
```

<h3> Rekurzivna eliminacija značajki </h3>

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

train, test = train_test_split(df, test_size=0.20, random_state=42)

X_train = train.drop('PRIJEVREMENI_RASKID',axis=1)
y_train = train['PRIJEVREMENI_RASKID']

rfe_rf = RFE(RandomForestClassifier(n_estimators=100,random_state=0,n_jobs=-1), n_features_to_select=15, step=1)
rfe_rf.fit(X_train, y_train);

print('Znacajke odabrane s RFE (koristeci slucajne sume):')
for i in list(train.columns.values[:-1][rfe_rf.get_support()]):
    print(' - ' + str(i))
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
train, test = train_test_split(df, test_size=0.20, random_state=42)

X_train = train.drop('PRIJEVREMENI_RASKID',axis=1)
y_train = train['PRIJEVREMENI_RASKID']

rfe_lr = RFE(LogisticRegression(solver='liblinear'), n_features_to_select=15, step=1)
rfe_lr.fit(X_train, y_train);

print('\nZnacajke odabrane s RFE (koristeci logisticku regresiju):')
for i in list(df.columns.values[:-1][rfe_lr.get_support()]):
    print(' - ' + str(i))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier

train, test = train_test_split(df, test_size=0.20, random_state=42)

X_train = train.drop('PRIJEVREMENI_RASKID',axis=1)
y_train = train['PRIJEVREMENI_RASKID']


rfecv_rf = RFECV(RandomForestClassifier(n_estimators=100,random_state=0,n_jobs=-1), 
                 step=1, 
                 cv=StratifiedKFold(n_splits=2),
                 scoring='f1') ## scoring!!!
rfecv_rf.fit(X_train, y_train);

print('Odabrane znacajke s RFECV (koristeci slucajne sume):')
for i in list(df.columns.values[:-1][rfecv_rf.get_support()]):
    print(' - ' + str(i))
```

Odabrana metoda je rekurzivna eliminacija značajki pomoću slučajnih šuma.

```python
features = df.columns.values[:-1][rfecv_rf.get_support()]
```

```python
X_train_ft = df[features]
X_train_ft['PRIJEVREMENI_RASKID'] = df['PRIJEVREMENI_RASKID']
X_train_ft.to_hdf('X_train_ft.h5', key = 'X_train')
```
