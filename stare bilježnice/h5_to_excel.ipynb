```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Ucitavanje train skupa
train = pd.read_hdf('train_set.h5')
```

```python
train.head()
```

```python
train['VALUTA_NAZIV'] = 'bla'
```

```python
train['VALUTA_NAZIV'].values[train['VALUTA'].values == 1] = 'HRK'
train['VALUTA_NAZIV'].values[train['VALUTA'].values == 2] = 'EUR'
train['VALUTA_NAZIV'].values[train['VALUTA'].values == 3] = 'USD'
train['VALUTA_NAZIV'].values[train['VALUTA'].values == 4] = 'CHF'
train['VALUTA_NAZIV'].values[train['VALUTA'].values == 5] = 'Ostalo'
```

```python
train.head()
```

```python
train['STAROST_KAT'] = 0
```

```python
lista = []
for i in train['STAROST']:
    if i<=10: lista.append('0-10')
    elif i<=20: lista.append('11-20')
    elif i<=30: lista.append('21-30')
    elif i<=40: lista.append('31-40')
    elif i<=50: lista.append('41-50')
    elif i<=60: lista.append('51-60')
    elif i<=70: lista.append('61-70')
    elif i<=80: lista.append('71-80')
    elif i<=90: lista.append('81-90')
    elif i<=100: lista.append('91-100')
    else: lista.append('više od 100')
train['STAROST_KAT'] =  lista
```

```python
train.head()
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('train_set.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
train.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
