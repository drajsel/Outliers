Kako cemo dati model na treniranje? Za svakog klijenta moze biti vise kredita i za taj kredit ima vise redaka informacija... Da li da gledamo prosjecne isplate, redovitost, postotak otplacenosti kredita? Za svaki redak imamo datum. Kako baratati s datumima? Bitna nam je godina zbog makroekonomskih podataka, a planirani_datum_zatvaranja mozemo promatrati drukcije, kao planirano trajanje kredita. Mozda da za svaki klijent i svaku njegovu partiju promotrimo isplate, i to svedemo na jedan redak.
